```python
def prepare_data_json():
    # 1. Define a function to clean text
    def clean_text(text):
        if not isinstance(text, str):
            return ""
        return (text
            .replace("\n", " ")
            .replace("\r", " ")
            .strip()
            .replace("  ", " ")
        )

    # 2. Create a structured document loader
    def load_documents_from_directory(directory, category):
        documents = []
        if not os.path.exists(directory):
            print(f"Warning: Directory {directory} does not exist")
            return documents
            
        for file in os.listdir(directory):
            try:
                if file.endswith(('.txt', '.md', '.json')):
                    with open(os.path.join(directory, file), 'r', encoding='utf-8') as f:
                        text = f.read()
                        documents.append({
                            "name": file,
                            "text": clean_text(text),
                            "category": category,
                            "source": directory
                        })
            except Exception as e:
                print(f"Error loading file {file}: {str(e)}")
        return documents

    # 3. Load all data sources with proper categorization
    data_json = []
    
    # Load phone conversations
    try:
        with open("../S05E01/phone.json", "r", encoding='utf-8') as f:
            phone_data = json.load(f)
            for key, value in phone_data.items():
                data_json.append({
                    "name": f"phone_conversation_{key}",
                    "text": clean_text(value),
                    "category": "phone_conversations",
                    "source": "phone.json"
                })
    except Exception as e:
        print(f"Error loading phone data: {str(e)}")

    # Load Zygfryd's notes
    data_json.extend(load_documents_from_directory("./zygfryd_notatnik", "zygfryd_notes"))
    
    # Load factory files
    data_json.extend(load_documents_from_directory("../pliki_z_fabryki/facts", "factory_facts"))
    data_json.extend(load_documents_from_directory("../pliki_z_fabryki/do-not-share", "factory_confidential"))
    
    # Load interrogation files
    data_json.extend(load_documents_from_directory("../S02E01/przesluchania", "interrogations"))
    
    # Load arxiv data
    data_json.extend(load_documents_from_directory("../S02E05/media_desc", "arxiv_media"))
    
    # 4. Add metadata and validate
    for doc in data_json:
        # Add length information
        doc['length'] = len(doc['text'])
        
        # Add timestamp if available
        if 'date' in doc['name']:
            try:
                # Extract date from filename if present
                date_str = re.search(r'\d{4}-\d{2}-\d{2}', doc['name'])
                if date_str:
                    doc['timestamp'] = date_str.group(0)
            except:
                pass

    # 5. Filter out empty or invalid documents
    data_json = [
        doc for doc in data_json 
        if doc['text'] and len(doc['text'].strip()) > 0
    ]

    # 6. Sort by category and name for better organization
    data_json.sort(key=lambda x: (x['category'], x['name']))

    # 7. Print summary statistics
    categories = {}
    for doc in data_json:
        cat = doc['category']
        categories[cat] = categories.get(cat, 0) + 1

    print("\nData Summary:")
    print(f"Total documents: {len(data_json)}")
    print("\nDocuments per category:")
    for cat, count in categories.items():
        print(f"{cat}: {count}")

    return data_json

# Use the function
data_json = prepare_data_json()

# Optional: Save prepared data to file for inspection
with open('prepared_data.json', 'w', encoding='utf-8') as f:
    json.dump(data_json, f, ensure_ascii=False, indent=2)
```

```python
# Example usage with your vector service
async def process_documents(data_json):
    points = []
    for item in data_json:
        try:
            doc = await text_splitter.document(
                text=item["text"],
                model="gpt-4o-mini",
                metadata={
                    "name": item["name"],
                    "category": item["category"],
                    "source": item["source"],
                    "length": item["length"]
                }
            )
            points.append(doc)
        except Exception as e:
            print(f"Error processing document {item['name']}: {str(e)}")
    
    return points
```

```python
COLLECTION_NAME = "S05E05"
async def initialize_data(openai: OpenAIService, 
                        vector_service: VectorService, 
                        text_splitter: TextSplitter) -> None:
    points = []
    for item in data_json:
        doc = await text_splitter.document(item["text"], "gpt-4o-mini", {"name": item["name"]})
        points.append(doc)
    print(json.dumps(points, indent=4, ensure_ascii=False))
    await vector_service.initialize_collection_with_data(COLLECTION_NAME, points)
```

```python
await initialize_data(openai, vector_service, text_splitter)
```

```python
for question_id, query in questions.items():
    print(f"\nProcessing question {question_id}: {query}")
    determine_raport_date = openai.completion(  
        messages=[
            {"role": "system", "content": f"""You are a helpful assistant that determines the name of the raport.
                                            Pick between {data_json}. just with name. If both are relevant, list them comma-separated. Write back with the name(s) and nothing else."""},
            {"role": "user", "content": query}
        ]
    )   
    print(determine_raport_date)
    page_names = determine_raport_date.choices[0].message.content.split(',') if determine_raport_date.choices[0].message.content else []
    page_names = [page_name.strip() for page_name in page_names]
    print(page_names)
```

```python
# Create filter
filter_dict = {
    "must": [
        {
            "key": "name",
            "match": {
                "value": page_name
            }
        } for page_name in page_names  # Use the page names from earlier
    ]
}

print(filter_dict)

# Perform search
search_results = await vector_service.perform_search(COLLECTION_NAME, query, filter_dict, 3)
print(search_results)
```

```python
# Check relevance
relevance_checks = []
for result in search_results:
    content = f"Query: {query}\nText: {result.payload.get('text', '')}"
    print(content)
    relevance_check = openai.completion(
        messages=[
            {"role": "system", "content": "You are a helpful assistant that determines if a given text is relevant to a query. Respond with 1 if relevant, 0 if not relevant."},
            {"role": "user", "content": content}
        ]
    )
    print(relevance_check)
    is_relevant = relevance_check.choices[0].message.content == "1"
    result_dict = dict(result)
    result_dict["is_relevant"] = is_relevant
    relevance_checks.append(result_dict)

relevant_results = [result for result in relevance_checks if result["is_relevant"]]
print(relevant_results)
```
